# LAB 4. GENETIC ALGORITHMS

Write a program that implements a genetic algorithm to solve the Knapsack Problem.

```python
import random
import numpy as np
np.random.seed(111)
random.seed(111)
```

```python
def fitness_function(ind, W_B, S):
    fitness_weight = 0
    fitness_benefit = 0
    for i in range(len(ind)):
        if ind[i] == 1:
            fitness_weight += ind[i] * W_B[0][i]
            fitness_benefit += ind[i] * W_B[1][i]
    if fitness_weight > S:
        return 0
    return fitness_benefit

def selection(population, fitness, tournament_size = 3):
    indices = np.random.choice(len(population), size=tournament_size)
    best_fitness = indices[np.argmax(fitness[indices])]
    return population[best_fitness]

def crossover(parent_1, parent_2):
    point = np.random.randint(1, len(parent_1) - 1)
    child_1 = np.concatenate([parent_1[:point], parent_2[point:]])
    child_2 = np.concatenate([parent_2[:point], parent_1[point:]])
    return child_1, child_2

def mutation(child):
    prob = 0.1
    if random.random() < prob:
        chromosome_replace = np.random.randint(0, len(child))
        if child[chromosome_replace] == 0:
            child[chromosome_replace] = 1
        else:
            child[chromosome_replace] = 0
    return child

def print_result(case_id, best_individual, W_B, benefit):
    selected_items = []
    for i in range(len(best_individual)):
        if best_individual[i] == 1:
            selected_items.append((W_B[0][i], W_B[1][i]))
    print(f"Case ID: {case_id}")
    print(f"M (number of items that could be replaced in the knapsack): {len(selected_items)}")
    print(f"Total benefit: {benefit}")
    for weight, benefit in selected_items:
        print(f"Benefit_Weight: ({weight}, {benefit})")

def GA_knapsack_problem(case_id, N_items, S, W_B):
    generations = 100
    population_size = 50
    population = np.random.randint(0, 2, size=(population_size, N_items))

    for generation in range(generations):
        fitness = np.array([fitness_function(ind, W_B, S) for ind in population])
        best_ind = np.argmax(fitness)
        new_population = []
        new_population.append(population[best_ind].copy())

        while len(new_population) < population_size:
            parent_1 = selection(population, fitness)
            parent_2 = selection(population, fitness)

            child_1, child_2 = crossover(parent_1, parent_2)

            child_1 = mutation(child_1)
            child_2 = mutation(child_2)

            new_population.append(child_1)
            if len(new_population) < population_size:
                new_population.append(child_2)

        population = np.array(new_population)
    fitness = np.array([fitness_function(ind, W_B, S) for ind in population])
    best_index = np.argmax(fitness)
    best_individual = population[best_index]

    benefit = 0
    for i in range(len(best_individual)):
        if best_individual[i] == 1:
            benefit += W_B[1][i]
    print_result(case_id, best_individual, W_B, benefit)
```

```python
С = 10
for case_id in range(1, C + 1):
    N_items = np.random.randint(10, 51)
    S = np.random.randint(50, 101)
    W_B = [
    np.random.randint(1, 20, size=N_items).tolist(),
    np.random.randint(1, 20, size=N_items).tolist()
    ]
    GA_knapsack_problem(case_id, N_items, S, W_B)
```
